```python
def A394696(nmax: int = 6) -> list[int]:
    a = []

    for n in range(2, nmax + 1):
        an = 0
        # Kanten vorab berechnen
        edges = [(i, j) for i in range(n) for j in range(i + 1, n)]
        E = len(edges)
        total_graphs = 1 << E

        for mask in range(total_graphs):
            degree = [0] * n
            adj = [0] * n

            # 1. Graphen aus der Bitmaske aufbauen
            for e, (u, v) in enumerate(edges):
                if (mask >> e) & 1:
                    adj[u] |= (1 << v)
                    adj[v] |= (1 << u)
                    degree[u] += 1
                    degree[v] += 1

            # 2. Prüfen: Keine isolierten Knoten (pr != 0)
            if 0 in degree:
                continue

            # 3. Bedingung prüfen: sum(Nachbarn) > di^2
            s_count = 0
            for i in range(n):
                sum_neighbor_degrees = 0
                neighbors = adj[i]

                # Bit-Trick: Iteriere nur über die tatsächlich gesetzten Bits
                while neighbors:
                    # Isoliere das niedrigste gesetzte Bit (z.B. 0b1010 -> 0b0010)
                    lowest_bit = neighbors & -neighbors
                    # Finde den Index des Knotens
                    j = lowest_bit.bit_length() - 1 

                    sum_neighbor_degrees += degree[j]
                    # Lösche das verarbeitete Bit
                    neighbors &= neighbors - 1

                # Entspricht dij/di - di > 0
                if sum_neighbor_degrees > degree[i] * degree[i]:
                    s_count += 1

            # 4. Count[s, 1] > n/2 (Ganzzahldivision!)
            if s_count > n // 2:
                an += 1

        a.append(an)
        print(f"n = {n} -> {an}")
    return a

if __name__ == '__main__':
    import time
    start_time = time.time()
    a = A394696()
    end_time = time.time()
    print("\nErgebnis a =", a)
    print(f"Berechnungszeit: {end_time - start_time:.4f} Sekunden")
```
