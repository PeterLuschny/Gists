```sage
def BellTransform(seq: list[int]) -> list[list[int]]:
    """
    Calculates the Bell-transform of a sequence. 
    The Bell-transform is a triangle related to the partial Bell polynomials.
    """
    N = len(seq) - 1
    T = [[0]*(n+1) for n in range(N+1)]
    T[0][0] = 1  # Base case
    for n in range(1, N+1):
        for k in range(1, n+1):
            T[n][k] = sum(binomial(n-1, j-1) * seq[j] * T[n-j][k-1]
                      for j in range(1, n-k+2))
    return T
```

Using a simple look-up function for the OEIS provided by SageMath. 

```sage
def OEIS(L: list[int], num=1, stop=False):
    s = " ".join(f"{v} " for v in L)
    R = oeis(s, max_results=num)
    A = [u.id() for u in R]
    if A == [] and not stop: 
        Y = [t for t in L[1:] if t != 0]
        R = OEIS(Y, 1, True) 
    return R if R else "[not found]"

def ShowTransform(seq: list[int]) -> None:
    """
    Shows the Bell-triangle of a sequence and its exponential transform.
    """
    T = BellTransform(seq)
    E = [sum(row) for row in T]
    print(f"Input sequence: {seq}")
    print(f"Exp-transform:  {E}")
    print(OEIS(seq))
    print(OEIS(E))
    print()
    print("Bell-triangle:")
    for row in T: print(row)
    print(OEIS([t for row in T for t in row]))
```
